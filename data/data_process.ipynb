# Cusine

### Processing cuisine

```python
category = pd.read_csv('category/All_Category_classified.csv')
```

```python
def category_processing(row):
    food_type = row['food_type']
    rest_type = row['restaurant_type']
    coun_type = row['country_type']
    types = [food_type, rest_type, coun_type]
    type_list = []
    for tp in types:
        if type(tp) is str:
            tp = tp.strip(';')
            subtype = tp.split(';')
            for tp in subtype:
                type_list.append(tp)
    for tp in type_list:
        row[tp] = 1
    return row
```

```python
city_ci_list = []
for city in city_list:
    city_temp = city
    for cat in category['category']:
        city_temp[cat] = 0
    city_cuisine = city_temp.apply(category_processing, axis = 1)
    city_ci_list.append(city_cuisine)
```

```python
for city in city_ci_list:
    name = city['city'][0]
    city.to_csv('city/{}_ci.csv'.format(name))
```

```python
type_total_df = pd.DataFrame(index = ['count','types'])
for city in city_ci_list:
    types_df = city.iloc[:,19:]
    name = city['city'][0]
    count = city.shape[0]
    var = round(sum(types_df.apply(lambda x : sum(x) ,axis = 0) > 0)/types_df.shape[1]*100,1)
    l = [count,var]
    type_total_df[name] = l
type_total_df.to_csv('city/pattern/type_counts.csv')
```

### sandwich & burger

```python
spb_df = pd.DataFrame(index = ['sandwich rate','pizza rate','burger rate','fast food rate','total rate'])
for city in city_ci_list:
    name = city['city'][0]
    count = city.shape[0]
    sandwich = city['Sandwiches']
    pizza = city['Pizza']
    burger = city['Burgers']
    ff = city['Fast Food']
    sr = sum(sandwich) / len(sandwich)
    pr = sum(pizza) / len(pizza)
    br = sum(burger) / len(burger)
    fr = sum(ff) / len(ff)
    sr = round(sr*100,1)
    br = round(br*100,1)
    pr = round(pr*100,1)
    fr = round(fr*100,1)
    tr = sr + br + pr + fr
    r = [sr,pr,br,fr,tr]
    spb_df[name] = r
#     print("{} sandwich rate: {}%, pizza rate: {}%, burger rate :{}%, total rate: {}%".format(name, sr, pr, br, tr))
spb_df.to_csv('city/pattern/fast_food.csv')
```

# Open time

### Longest and shortest open time

```python
import time
vegas = pd.read_csv("city/las_vegas.csv")
madison = pd.read_csv("city/Madison.csv")
phoenix = pd.read_csv("city/Phoenix.csv")
cleveland = pd.read_csv("city/Cleveland.csv")
charlotte = pd.read_csv("city/Charlotte.csv")
montreal = pd.read_csv("city/Montreal.csv")
toronto = pd.read_csv("city/Toronto.csv")
city_list = [vegas, madison, phoenix, cleveland, charlotte, montreal, toronto]
```

```python
def time_span(timerange):
    if timerange != 'None':
        try:
            timerange = timerange.split('-')
            start, end = (time.strptime(timerange[0], "%H:%M"), time.strptime(timerange[1], "%H:%M"))
            start_hour, start_min = start.tm_hour,start.tm_min
            end_hour, end_min = end.tm_hour, end.tm_min
            st1,st2,et1,et2 = start_hour, start_min, end_hour, end_min
            if et1 <= st1 : et1 += 24
            timespan = et1 - st1 + (et2 - st2) / 60
            return timespan
        except:
            return None
    else: return None
```

```python
def city_dataframe(time_func):
    city_df = pd.DataFrame()
    for city in city_list:
        name = city['city'][0]
        city_time = city.iloc[:,-7:]
        t = city_time.applymap(time_func)
        city_df[name] = t.apply(lambda x: x.mean(), axis = 0)
    city_df.loc['Average'] = city_df.apply(lambda x: x.mean())
    city_df['Average'] = city_df.apply(lambda x: x.mean(), axis = 1)
    return city_df
```

```python
round(city_dataframe(time_span),2).to_csv('city/pattern/open_duration.csv')
```

### 24h rate

```python
def time_24(timerange):
    if timerange != 'None':
        try:
            timerange = timerange.split('-')
            start, end = (time.strptime(timerange[0], "%H:%M"), time.strptime(timerange[1], "%H:%M"))
            start_hour, start_min = start.tm_hour,start.tm_min
            end_hour, end_min = end.tm_hour, end.tm_min
            st1,st2,et1,et2 = start_hour, start_min, end_hour, end_min
            if st1 == et1 and st2 == et2: return 1 
            else: return 0
        except:
            return None
    else: return None
```

```python
round(city_dataframe(time_24)*100,2).to_csv('city/pattern/24hour_rate.csv')
```

### Earliest and latest open time

```python
def time_early(timerange):
    if timerange != 'None':
        try:
            timerange = timerange.split('-')
            start, end = (time.strptime(timerange[0], "%H:%M"), time.strptime(timerange[1], "%H:%M"))
            start_hour, start_min = start.tm_hour,start.tm_min
            end_hour, end_min = end.tm_hour, end.tm_min
            st1,st2,et1,et2 = start_hour, start_min, end_hour, end_min
            return st1 + st2/60
        except:
            return None
    else: return None
```

```python
round(city_dataframe(time_early),2).to_csv('city/pattern/early.csv')
```

```python
def time_late(timerange):
    if timerange != 'None':
        try:
            timerange = timerange.split('-')
            start, end = (time.strptime(timerange[0], "%H:%M"), time.strptime(timerange[1], "%H:%M"))
            start_hour, start_min = start.tm_hour,start.tm_min
            end_hour, end_min = end.tm_hour, end.tm_min
            st1,st2,et1,et2 = start_hour, start_min, end_hour, end_min
            if et1 == 0: et1 = 24
            return et1 + et2/60
        except:
            return None
    else: return None
```

```python
round(city_dataframe(time_late),2).to_csv('city/pattern/late.csv')
```

# Cuisine X Rating

### Highest rating cuisine(rest/coun/food)

```python
def cuisine_cross_pattern(cross_att):
    city = city_ci_list[0]
    city = city.iloc[:,21:50]
    cuisine_cross_df = pd.DataFrame(index = city.columns)
    for city in city_ci_list:
        name = city['city'][0]
        cross = city[cross_att]
        city = city.iloc[:,21:50]
        cuisine_cross_df[name] = 0
        for cat in city:
            cuisine_cross = city[cat] * cross
            cuisine_cross = cuisine_cross[cuisine_cross != 0]
            cuisine_cross = cuisine_cross.mean()
            cuisine_cross_df[name].loc[cat] = cuisine_cross
    cuisine_cross_df.loc['Sum'] = cuisine_cross_df.apply(lambda x: x.mean())
    return cuisine_cross_df
```

```python
temp = round(cuisine_cross_pattern('stars'),2)
temp.to_csv('city/pattern/cuisine_X_rating.csv')
```

### Most popular cuisine(r/c/f)

```python
temp = round(cuisine_cross_pattern('review_count'),2)
temp.to_csv('city/pattern/cuisine_X_review.csv')
```
